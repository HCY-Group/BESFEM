## Plotting Using MFEM's PyGLViS

More information: https://github.com/GLVis/pyglvis

```python
# Pip installs pyglvis (and pymfem) if in a colab notebook. Can comment out if not required
%pip install --quiet mfem==4.8.0 glvis
from glvis import glvis, GlvisData
import mfem.ser as mfem
import numpy as np
```

```python
mesh = mfem.Mesh()
mesh.Load("../bin/pmesh")
x = mfem.GridFunction(mesh,"../bin/psi")
```

```python
# Plot the mesh + solution (all GLVis keys and mouse commands work)
g = glvis((mesh, x))
g
```
